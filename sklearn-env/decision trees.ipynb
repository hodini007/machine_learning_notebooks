```python
df=pd.read_csv('titanic.csv')
df
```

```python
target =df.Survived
target
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived'],axis='columns',inplace=True)
```

```python
df
```

```python
input=df
```

```python
input.Sex=input.Sex.map({'male':1,'female':2})
```

```python
input
```

```python
input.Age = input.Age.fillna(input.Age.mean())
input
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion='gini')
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
